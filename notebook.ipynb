## Anomaly detection using isolation forest

```python
import pandas as pd
df=pd.read_csv('healthcare.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df.iloc[:,0],df.iloc[:,1])
```

```python
from sklearn.ensemble import IsolationForest
```

```python
clf=IsolationForest(contamination=0.2)
clf.fit(df)
predictions=clf.predict(df)
```

```python
predictions
```

```python
import numpy as np 
index=np.where(predictions<0)
index
```

```python
x=df.values
```

```python
plt.scatter(df.iloc[:,0],df.iloc[:,1])
plt.scatter(x[index,0],x[index,1],edgecolors='r')
```

## Anamoly detection using DBscan clustering

```python
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import make_circles
```

```python
X,y=make_circles(n_samples=750,factor=0.3,noise=0.1)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
dbscan=DBSCAN(eps=0.10)
dbscan.fit_predict(X)
```

```python
dbscan.labels_
```

```python
plt.scatter(X[:,0],X[:,1],c=dbscan.labels_)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

## Anomaly detection using local outlier factor (lof)

```python
#internaly it uses the knearest neighbor- detects the local density
```

```python
#distance is more - then density is high
#distance is less - then density is low

```

```python
from sklearn.neighbors import LocalOutlierFactor
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import make_circles

```

```python
X,y=make_circles(n_samples=750,factor=0.3,noise=0.1)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
lof=LocalOutlierFactor(n_neighbors=5)
lof.fit_predict(X)
```

```python
lof.negative_outlier_factor_
```

```python
plt.scatter(X[:,0],X[:,1],c=lof.negative_outlier_factor_)
```
